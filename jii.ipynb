```python
from sklearn import linear_model
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error as mae,median_absolute_error as mee,classification_report as cr,accuracy_score as ac
```

```python
x=df[['ph', 'Hardness', 'Solids', 'Chloramines', 'Sulfate', 'Conductivity','Organic_carbon', 'Trihalomethanes', 'Turbidity']]
y=df[['Potability']]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=10)
```

```python
from sklearn.naive_bayes import GaussianNB  
classifier1 = GaussianNB()  
classifier1.fit(x_train, y_train) 
y_pred3= classifier1.predict(x_test)  
print(ac(y_test,y_pred3)*100)
```

```python
import pickle
pickle.dump(classifier1,open('model1.pkl','wb'))
model=pickle.load(open('model1.pkl','rb'))
```
